```python
    graph = None
    import functools, inspect, traitlets
```

```python
    def load_ipython_extension(shell):
        import networkx
        global graph
        graph = graph or networkx.DiGraph()
        shell.user_ns['GRAPH'] = shell.user_ns.get('GRAPH', graph)
        traitlets.dlink = inspect.unwrap(traitlets.dlink)
        traitlets.dlink = dlink
```

```python
    def unload_ipython_extension(shell):
        traitlets.HasTraits.observe = inspect.unwrap(traitlets.HasTraits.observe)
        traitlets.HasTraits.observe = observe
```

```python
    @functools.wraps(traitlets.dlink)
    def dlink(source, target, transform=None):
        graph.add_edge(source, target, callable=transform)
        return inspect.unwrap(traitlets.dlink)(source, target, transform)
```

```python
    @functools.wraps(traitlets.HasTraits.observe)
    def observe(self, handler, names=traitlets.All, type='change',):
        if isinstance(names, str):
            names = names.split()
        for name in names:
            graph.add_edge((self, name), handler)
        return inspect.unwrap(traitlets.HasTraits.observe)(self, handler, names=names, type=type)
```
