```python
    @contextlib.contextmanager
    def swap_sys(sys_modules):
        import collections
        sys.modules = collections.ChainMap(sys_modules, sys.modules)
        sys.modules['__main__'].__getattribute__('__name__')
        yield sys.modules
        sys.modules = sys.modules.maps[-1]
```

```python
    def static_types(file, **dict):
        for _0, _1 in mypy.main.build.build(
            [mypy.main.BuildSource(file, '__main__', None)], mypy.main.Options()).files.items():
            loader = importlib.machinery.SourceFileLoader(_0, _1.path)
            dict[_0] = module = importlib.util.module_from_spec(importlib.machinery.ModuleSpec(_0, importlib.util.LazyLoader(loader)))
            module.__loader__.exec_module(module)
        with swap_sys(dict): dir(dict['__main__'])
        return dict['__main__']
        
```
